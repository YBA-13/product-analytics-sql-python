```python
import pandas as pd

data = {
    "order_id": [1, 2, 3, 4, 5, 6, 7],
    "customer_id": [101, 102, 101, 103, 104, 102, 101],
    "category": ["Electronics", "Home", "Electronics", "Sports", "Home", "Electronics", "Sports"],
    "product_name": ["Headphones", "Lamp", "Mouse", "Dumbbell", "Chair", "Keyboard", "Yoga Mat"],
    "quantity": [2, 1, 1, 2, 1, 1, 1],
    "price": [15000, 8000, 5000, 12000, 25000, 10000, 7000]
}

df = pd.DataFrame(data)
df
```

```python
revenue_by_category = (
    df.groupby("category")
      .apply(lambda x: (x["price"] * x["quantity"]).sum())
      .reset_index(name="total_revenue")
      .sort_values("total_revenue", ascending=False)
)

revenue_by_category
```

```python
import matplotlib.pyplot as plt

revenue_by_category.plot(
    x="category",
    y="total_revenue",
    kind="bar",
    figsize=(8,5)
)

plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=0)
plt.show()
```

```python
revenue_by_customer = (
    df.groupby("customer_id")
      .apply(lambda x: (x["price"] * x["quantity"]).sum())
      .reset_index(name="total_revenue")
      .sort_values("total_revenue", ascending=False)
)

revenue_by_customer
```

```python
revenue_by_customer.plot(
    x="customer_id",
    y="total_revenue",
    kind="bar",
    figsize=(8,5)
)

plt.title("Revenue by Customer")
plt.xlabel("Customer ID")
plt.ylabel("Revenue")
plt.xticks(rotation=0)
plt.show()
```
